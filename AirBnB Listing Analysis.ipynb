```python
# filtering data down to rows where the city is Paris, 
# keeping only the columns ‘host_since’, ‘neighbourhood’, ‘city’, ‘accommodates’, and ‘price’

paris_listings = (
                 listings
                 .query("city == 'Paris'")
                 .loc[:,["host_since", "neighbourhood","city", "accommodates", "price"]]
                 )

paris_listings.info()
```

## Preparing the data for visualization

```python
# creating  a table named paris_listings_neighbourhood that groups Paris listings by 'neighbourhood' 
# calculating the mean price 
# sorting price from low to high

paris_listings_neighbourhood = (
                                paris_listings
                                .groupby("neighbourhood")
                                .agg({"price":"mean"})
                                .sort_values("price")
                               )

paris_listings_neighbourhood.tail()
```

```python
# creating a table named paris_listings_accomodations 
# filtering down to the most expensive neighborhood,
# grouping by the ‘accommodations’ column
# adding the mean price for each value of ‘accommodates’
# sorting price from low to high

paris_listings_accomodations = (
                                paris_listings
                                .query("neighbourhood == 'Elysee'")
                                .groupby("accommodates")
                                .agg({"price":"mean"})
                                .sort_values("price")
                               )

paris_listings_accomodations.head()
```

```python
# creating a table called paris_listings_over_time grouped by the ‘host_since’ year, 
# calculating the average price
# counting of rows representing the number of new hosts

paris_listings_over_time = (
                            paris_listings
                            .set_index("host_since")
                            .resample("Y")
                            .agg({"neighbourhood":"count","price":"mean"})
                            )

paris_listings_over_time.head()
```

## Visualizing the data and summarizing findings

```python
# creating a horizontal bar chart of the average price by neighborhood in Paris, 
# adding title and axis labels

(
paris_listings_neighbourhood
.plot
.barh(
     title = "Average Listing Price by Paris Neighbourhood",
     xlabel = " Price Per Night (Euros)",
      ylabel = "Neighbourhood"
    )
);
```

```python
# importing seaborn package
# removing the legend

import seaborn as sns

(
paris_listings_neighbourhood
.plot
.barh(
     title = "Average Listing Price by Paris Neighbourhood",
     xlabel = " Price Per Night (Euros)",
      ylabel = "Neighbourhood",
     legend = None
    )
);

sns.despine()
```

```python
# creating a horizontal bar chart of the average price by 'accommodates' in Paris' most expensive neighborhood 
# adding title and axis labels

(
paris_listings_accomodations
.plot
.barh(
     title = "Average Listing Price by Accomodations Number",
     xlabel = " Price Per Night (Euros)",
      ylabel = "Accomodation Capacity",
     legend = None
    )
);

sns.despine()
```

```python
# creating a line chart to show the count of new hosts over time


paris_listings_over_time["neighbourhood"].plot(
                                               title = "New AirBnB Hosts in Paris Over Time",
                                                ylabel = "New Hosts"
                                                );


sns.despine()
```

```python
# creating a line chart to show average price over time

paris_listings_over_time["price"].plot(
                                               title = "Average AirBnB Price in Paris Over Time",
                                                ylabel = "Average Price (Euros)"
                                                );


sns.despine()
```
